```python
import pandas as pd
import json
```

```python
# -------------------------------
ruta_json = "/content/TelecomX_Data.json"

with open(ruta_json, "r") as f:
    data = json.load(f)

# Convertir JSON a DataFrame (expandiendo los campos anidados)
df = pd.json_normalize(data)

print("‚úÖ Dataset cargado correctamente")
print("N√∫mero de filas:", df.shape[0])
print("N√∫mero de columnas:", df.shape[1])
```

```python
# Valores nulos
print("üîé Valores nulos por columna:")
print(df.isnull().sum())

# Duplicados
print("\nüìå Registros duplicados:", df.duplicated().sum())

# Categor√≠as √∫nicas en variables clave
cat_cols = [
    "Churn", "customer.gender", "customer.Partner", "customer.Dependents",
    "phone.PhoneService", "internet.InternetService",
    "account.Contract", "account.PaperlessBilling", "account.PaymentMethod"
]

for col in cat_cols:
    print(f"\nüìå {col}: {df[col].unique()}")

# Revisar columnas num√©ricas
print("\nüìä Tipos de datos num√©ricos:")
print(df[["customer.tenure", "account.Charges.Monthly", "account.Charges.Total"]].dtypes)
```

#üîß Transformaci√≥n

```python
# Copia del DataFrame
df_clean = df.copy()

# 1. Convertir TotalCharges a num√©rico
df_clean["account.Charges.Total"] = pd.to_numeric(df_clean["account.Charges.Total"], errors="coerce")

# 2. Reemplazar valores vac√≠os en Churn por NaN
df_clean["Churn"] = df_clean["Churn"].replace("", pd.NA)

# 3. Homogeneizar categor√≠as ("No internet service"/"No phone service" ‚Üí "No")
cols_to_fix = [
    "phone.MultipleLines", "internet.OnlineSecurity", "internet.OnlineBackup",
    "internet.DeviceProtection", "internet.TechSupport",
    "internet.StreamingTV", "internet.StreamingMovies"
]

for col in cols_to_fix:
    df_clean[col] = df_clean[col].replace({"No internet service": "No", "No phone service": "No"})

# 4. Verificar
print("üîé Valores nulos tras limpieza:")
print(df_clean.isnull().sum())
```

```python
# Nueva columna con facturaci√≥n diaria (aprox. 30 d√≠as)
df_clean["Cuentas_Diarias"] = df_clean["account.Charges.Monthly"] / 30

print("‚úÖ Columna 'Cuentas_Diarias' creada")
print(df_clean[["account.Charges.Monthly", "Cuentas_Diarias"]].head())
```

```python
print(df_clean.columns.tolist())
```

```python
# 1) Revisar nombres originales
print("üîé Nombres originales de columnas:")
print(list(df.columns))
```

```python
# Normalizar: quitar espacios, min√∫sculas, reemplazar "." y " " por "_"
df_clean.columns = (
    df_clean.columns
      .str.strip()                # elimina espacios al inicio/fin
      .str.lower()                # pasa todo a min√∫scula
      .str.replace(" ", "_")      # espacios ‚Üí _
      .str.replace(".", "_")      # puntos ‚Üí _
)

print("‚úÖ Columnas normalizadas:")
print(df_clean.columns.tolist())

```

```python
# ============================
# 1. Restaurar nombres originales + columna extra
# ============================
original_cols = [
    'customerID', 'Churn', 'customer.gender', 'customer.SeniorCitizen',
    'customer.Partner', 'customer.Dependents', 'customer.tenure',
    'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
    'internet.OnlineSecurity', 'internet.OnlineBackup',
    'internet.DeviceProtection', 'internet.TechSupport',
    'internet.StreamingTV', 'internet.StreamingMovies',
    'account.Contract', 'account.PaperlessBilling', 'account.PaymentMethod',
    'account.Charges.Monthly', 'account.Charges.Total',
    'Cuentas_Diarias'   # üëà a√±adida
]

# Forzar asignaci√≥n
df_clean.columns = original_cols

# ============================
# 2. Normalizar nombres
# ============================
df_clean.columns = (
    df_clean.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(".", "_")
)

print("‚úÖ Columnas normalizadas:")
print(df_clean.columns.tolist())

# ============================
# 3. Traducci√≥n a espa√±ol amigable
# ============================
col_translation = {
    "customerid": "id_cliente",
    "churn": "evasion",
    "customer_gender": "genero",
    "customer_seniorcitizen": "adulto_mayor",
    "customer_partner": "pareja",
    "customer_dependents": "dependientes",
    "customer_tenure": "meses_contrato",
    "phone_phoneservice": "servicio_telefonico",
    "phone_multiplelines": "multiples_lineas",
    "internet_internetservice": "servicio_internet",
    "internet_onlinesecurity": "seguridad_online",
    "internet_onlinebackup": "respaldo_online",
    "internet_deviceprotection": "proteccion_dispositivo",
    "internet_techsupport": "soporte_tecnico",
    "internet_streamingtv": "streaming_tv",
    "internet_streamingmovies": "streaming_peliculas",
    "account_contract": "tipo_contrato",
    "account_paperlessbilling": "factura_electronica",
    "account_paymentmethod": "metodo_pago",
    "account_charges_monthly": "cargo_mensual",
    "account_charges_total": "cargo_total",
    "cuentas_diarias": "cargo_diario"
}

df_final = df_clean.rename(columns=col_translation)

# ============================
# 4. Verificaci√≥n
# ============================
print("\nüìã Columnas en espa√±ol listas para an√°lisis:")
print(df_final.columns.tolist())

print("\nüìä Vista previa:")
print(df_final.head())
```

```python
# Copiamos el df_final
df_ml = df_final.copy()

# 1. Convertir S√≠/No ‚Üí 1/0
binary_map = {"Yes": 1, "No": 0}
for col in ["evasion", "pareja", "dependientes", "servicio_telefonico",
            "multiples_lineas", "seguridad_online", "respaldo_online",
            "proteccion_dispositivo", "soporte_tecnico",
            "streaming_tv", "streaming_peliculas", "factura_electronica"]:
    df_ml[col] = df_ml[col].map(binary_map)

# 2. Convertir g√©nero a binario (Female=0, Male=1)
df_ml["genero"] = df_ml["genero"].map({"Female": 0, "Male": 1})

# 3. Variables categ√≥ricas ‚Üí One-Hot Encoding
df_ml = pd.get_dummies(df_ml, columns=["tipo_contrato", "metodo_pago", "servicio_internet"],
                       drop_first=True)

print(df_ml.head())
```

# üìä Informe Final - An√°lisis de Evasi√≥n de Clientes (Churn) en Telecom X

---

## üîπ Introducci√≥n

Telecom X enfrenta un desaf√≠o importante: una **alta tasa de evasi√≥n de clientes (Churn)**.  
El objetivo de este an√°lisis fue **explorar y entender los factores que influyen en la cancelaci√≥n de servicios**, con el fin de generar insights estrat√©gicos que ayuden a la empresa a **retener clientes y reducir p√©rdidas**.

El proyecto se desarroll√≥ en varias etapas: **importaci√≥n, limpieza, transformaci√≥n, an√°lisis exploratorio y conclusiones estrat√©gicas**.

---

## üîπ Limpieza y Tratamiento de Datos

1. **Carga de datos**: Se importaron los registros en formato JSON y se transformaron a un DataFrame de Pandas.  
2. **Exploraci√≥n inicial**: El dataset conten√≠a **7.267 clientes y 21 variables** (demogr√°ficas, servicios contratados y datos de facturaci√≥n).  
3. **Problemas encontrados**:
   - Valores nulos en `Churn` y `account.Charges.Total`.
   - Variables con formatos inconsistentes (`Yes/No`, `Male/Female`).
   - Columnas con nombres t√©cnicos poco claros.  
4. **Tratamiento aplicado**:
   - Eliminaci√≥n y correcci√≥n de nulos e inconsistencias.
   - Conversi√≥n de `account.Charges.Total` a num√©rico.
   - Creaci√≥n de una nueva variable: **`cargo_diario`** (`cargo_mensual / 30`).
   - Normalizaci√≥n y traducci√≥n de columnas al espa√±ol (ej. `customer.gender` ‚Üí `genero`, `Churn` ‚Üí `evasion`).

Con esto, se obtuvo un dataset **limpio, coherente y listo para an√°lisis**.

---

## üîπ An√°lisis Exploratorio de Datos (EDA)

### 1. Estad√≠sticas descriptivas
- **Cargo mensual**: promedio ‚âà 64 USD, rango de 18 a 118 USD.
- **Meses de contrato (tenure)**: clientes con 0 a 72 meses, con media alrededor de 32 meses.
- **Cargo total**: gran variaci√≥n, de pocos d√≥lares hasta m√°s de 8.000 USD.

### 2. Distribuci√≥n de evasi√≥n (Churn)
- Aproximadamente **26% de los clientes cancelaron** el servicio.
- La mayor√≠a (74%) permanecen activos.

### 3. Evasi√≥n por variables categ√≥ricas
- **Tipo de contrato**: los clientes con contrato *‚Äúmonth-to-month‚Äù* tienen mayor evasi√≥n que los de ‚Äúone year‚Äù o ‚Äútwo year‚Äù.
- **M√©todo de pago**: los que usan *‚ÄúElectronic check‚Äù* presentan m√°s evasi√≥n en comparaci√≥n con pagos autom√°ticos.
- **G√©nero**: no se encontraron diferencias significativas entre hombres y mujeres.

### 4. Evasi√≥n por variables num√©ricas
- **Meses de contrato**: clientes con menor antig√ºedad son m√°s propensos a cancelar.
- **Cargo mensual**: quienes pagan m√°s caro tienden a evadir en mayor proporci√≥n.
- **Cargo total**: clientes con bajo gasto acumulado muestran m√°s evasi√≥n, lo que refuerza la idea de que cancelan temprano en su ciclo.

---

## üîπ Conclusiones e Insights

1. **Contratos cortos = m√°s evasi√≥n**  
   Los clientes con contrato mensual muestran una clara propensi√≥n a cancelar.  

2. **M√©todos de pago influyen**  
   El pago por *‚ÄúElectronic check‚Äù* se asocia con mayor churn, posiblemente por ser menos c√≥modo que los pagos autom√°ticos.  

3. **Clientes nuevos son m√°s vulnerables**  
   La mayor√≠a de los que cancelan tienen pocos meses de contrato.  

4. **Altos cargos mensuales = mayor churn**  
   El precio elevado es un factor que impulsa la evasi√≥n.  

---

## üîπ Recomendaciones

1. **Incentivar contratos de largo plazo**  
   Ofrecer descuentos y beneficios para clientes que cambien de ‚Äúmonth-to-month‚Äù a ‚Äúone year‚Äù o ‚Äútwo year‚Äù.

2. **Fomentar pagos autom√°ticos**  
   Dar beneficios a clientes que pasen de *Electronic check* a *Bank transfer* o *Credit card (automatic)*.  

3. **Programas de fidelizaci√≥n para clientes nuevos**  
   Primeros 6 meses son cr√≠ticos: ofrecer promociones, soporte especial y beneficios exclusivos para reducir cancelaciones tempranas.  

4. **Revisi√≥n de precios y segmentaci√≥n**  
   Analizar clientes con cargos mensuales altos y ofrecer planes m√°s flexibles para mejorar la retenci√≥n.
