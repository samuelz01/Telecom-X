#📌 Extracción

```python
import pandas as pd
import json
```

```python
# -------------------------------
ruta_json = "/content/TelecomX_Data.json"

with open(ruta_json, "r") as f:
    data = json.load(f)

# Convertir JSON a DataFrame (expandiendo los campos anidados)
df = pd.json_normalize(data)

print("✅ Dataset cargado correctamente")
print("Número de filas:", df.shape[0])
print("Número de columnas:", df.shape[1])
```

```python
# Primeras filas
print("\n📊 Primeras filas del dataset:")
display(df.head())

# Información de columnas y tipos de datos
print("\n📋 Información general del DataFrame:")
df.info()

# Tipos de datos
print("\n🔎 Tipos de datos por columna:")
print(df.dtypes)
```

```python
# 1. Explorar columnas y tipos de datos
print("📊 Información general del DataFrame:")
print(df.info())

print("\n🔎 Primeras filas del dataset:")
display(df.head())

print("\n📋 Tipos de datos por columna:")
print(df.dtypes)
```

```python
# Valores nulos
print("🔎 Valores nulos por columna:")
print(df.isnull().sum())

# Duplicados
print("\n📌 Registros duplicados:", df.duplicated().sum())

# Categorías únicas en variables clave
cat_cols = [
    "Churn", "customer.gender", "customer.Partner", "customer.Dependents",
    "phone.PhoneService", "internet.InternetService",
    "account.Contract", "account.PaperlessBilling", "account.PaymentMethod"
]

for col in cat_cols:
    print(f"\n📌 {col}: {df[col].unique()}")

# Revisar columnas numéricas
print("\n📊 Tipos de datos numéricos:")
print(df[["customer.tenure", "account.Charges.Monthly", "account.Charges.Total"]].dtypes)
```

#🔧 Transformación

```python
# Copia del DataFrame
df_clean = df.copy()

# 1. Convertir TotalCharges a numérico
df_clean["account.Charges.Total"] = pd.to_numeric(df_clean["account.Charges.Total"], errors="coerce")

# 2. Reemplazar valores vacíos en Churn por NaN
df_clean["Churn"] = df_clean["Churn"].replace("", pd.NA)

# 3. Homogeneizar categorías ("No internet service"/"No phone service" → "No")
cols_to_fix = [
    "phone.MultipleLines", "internet.OnlineSecurity", "internet.OnlineBackup",
    "internet.DeviceProtection", "internet.TechSupport",
    "internet.StreamingTV", "internet.StreamingMovies"
]

for col in cols_to_fix:
    df_clean[col] = df_clean[col].replace({"No internet service": "No", "No phone service": "No"})

# 4. Verificar
print("🔎 Valores nulos tras limpieza:")
print(df_clean.isnull().sum())
```

```python
# Nueva columna con facturación diaria (aprox. 30 días)
df_clean["Cuentas_Diarias"] = df_clean["account.Charges.Monthly"] / 30

print("✅ Columna 'Cuentas_Diarias' creada")
print(df_clean[["account.Charges.Monthly", "Cuentas_Diarias"]].head())
```

```python
print(df_clean.columns.tolist())
```

```python
# 1) Revisar nombres originales
print("🔎 Nombres originales de columnas:")
print(list(df.columns))
```

```python
# Normalizar: quitar espacios, minúsculas, reemplazar "." y " " por "_"
df_clean.columns = (
    df_clean.columns
      .str.strip()                # elimina espacios al inicio/fin
      .str.lower()                # pasa todo a minúscula
      .str.replace(" ", "_")      # espacios → _
      .str.replace(".", "_")      # puntos → _
)

print("✅ Columnas normalizadas:")
print(df_clean.columns.tolist())

```

```python
# ============================
# 1. Restaurar nombres originales + columna extra
# ============================
original_cols = [
    'customerID', 'Churn', 'customer.gender', 'customer.SeniorCitizen',
    'customer.Partner', 'customer.Dependents', 'customer.tenure',
    'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
    'internet.OnlineSecurity', 'internet.OnlineBackup',
    'internet.DeviceProtection', 'internet.TechSupport',
    'internet.StreamingTV', 'internet.StreamingMovies',
    'account.Contract', 'account.PaperlessBilling', 'account.PaymentMethod',
    'account.Charges.Monthly', 'account.Charges.Total',
    'Cuentas_Diarias'   # 👈 añadida
]

# Forzar asignación
df_clean.columns = original_cols

# ============================
# 2. Normalizar nombres
# ============================
df_clean.columns = (
    df_clean.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(".", "_")
)

print("✅ Columnas normalizadas:")
print(df_clean.columns.tolist())

# ============================
# 3. Traducción a español amigable
# ============================
col_translation = {
    "customerid": "id_cliente",
    "churn": "evasion",
    "customer_gender": "genero",
    "customer_seniorcitizen": "adulto_mayor",
    "customer_partner": "pareja",
    "customer_dependents": "dependientes",
    "customer_tenure": "meses_contrato",
    "phone_phoneservice": "servicio_telefonico",
    "phone_multiplelines": "multiples_lineas",
    "internet_internetservice": "servicio_internet",
    "internet_onlinesecurity": "seguridad_online",
    "internet_onlinebackup": "respaldo_online",
    "internet_deviceprotection": "proteccion_dispositivo",
    "internet_techsupport": "soporte_tecnico",
    "internet_streamingtv": "streaming_tv",
    "internet_streamingmovies": "streaming_peliculas",
    "account_contract": "tipo_contrato",
    "account_paperlessbilling": "factura_electronica",
    "account_paymentmethod": "metodo_pago",
    "account_charges_monthly": "cargo_mensual",
    "account_charges_total": "cargo_total",
    "cuentas_diarias": "cargo_diario"
}

df_final = df_clean.rename(columns=col_translation)

# ============================
# 4. Verificación
# ============================
print("\n📋 Columnas en español listas para análisis:")
print(df_final.columns.tolist())

print("\n📊 Vista previa:")
print(df_final.head())
```

```python
# Copiamos el df_final
df_ml = df_final.copy()

# 1. Convertir Sí/No → 1/0
binary_map = {"Yes": 1, "No": 0}
for col in ["evasion", "pareja", "dependientes", "servicio_telefonico",
            "multiples_lineas", "seguridad_online", "respaldo_online",
            "proteccion_dispositivo", "soporte_tecnico",
            "streaming_tv", "streaming_peliculas", "factura_electronica"]:
    df_ml[col] = df_ml[col].map(binary_map)

# 2. Convertir género a binario (Female=0, Male=1)
df_ml["genero"] = df_ml["genero"].map({"Female": 0, "Male": 1})

# 3. Variables categóricas → One-Hot Encoding
df_ml = pd.get_dummies(df_ml, columns=["tipo_contrato", "metodo_pago", "servicio_internet"],
                       drop_first=True)

print(df_ml.head())
```

# 📊 Informe Final - Análisis de Evasión de Clientes (Churn) en Telecom X

---

## 🔹 Introducción

Telecom X enfrenta un desafío importante: una **alta tasa de evasión de clientes (Churn)**.  
El objetivo de este análisis fue **explorar y entender los factores que influyen en la cancelación de servicios**, con el fin de generar insights estratégicos que ayuden a la empresa a **retener clientes y reducir pérdidas**.

El proyecto se desarrolló en varias etapas: **importación, limpieza, transformación, análisis exploratorio y conclusiones estratégicas**.

---

## 🔹 Limpieza y Tratamiento de Datos

1. **Carga de datos**: Se importaron los registros en formato JSON y se transformaron a un DataFrame de Pandas.  
2. **Exploración inicial**: El dataset contenía **7.267 clientes y 21 variables** (demográficas, servicios contratados y datos de facturación).  
3. **Problemas encontrados**:
   - Valores nulos en `Churn` y `account.Charges.Total`.
   - Variables con formatos inconsistentes (`Yes/No`, `Male/Female`).
   - Columnas con nombres técnicos poco claros.  
4. **Tratamiento aplicado**:
   - Eliminación y corrección de nulos e inconsistencias.
   - Conversión de `account.Charges.Total` a numérico.
   - Creación de una nueva variable: **`cargo_diario`** (`cargo_mensual / 30`).
   - Normalización y traducción de columnas al español (ej. `customer.gender` → `genero`, `Churn` → `evasion`).

Con esto, se obtuvo un dataset **limpio, coherente y listo para análisis**.

---

## 🔹 Análisis Exploratorio de Datos (EDA)

### 1. Estadísticas descriptivas
- **Cargo mensual**: promedio ≈ 64 USD, rango de 18 a 118 USD.
- **Meses de contrato (tenure)**: clientes con 0 a 72 meses, con media alrededor de 32 meses.
- **Cargo total**: gran variación, de pocos dólares hasta más de 8.000 USD.

### 2. Distribución de evasión (Churn)
- Aproximadamente **26% de los clientes cancelaron** el servicio.
- La mayoría (74%) permanecen activos.

### 3. Evasión por variables categóricas
- **Tipo de contrato**: los clientes con contrato *“month-to-month”* tienen mayor evasión que los de “one year” o “two year”.
- **Método de pago**: los que usan *“Electronic check”* presentan más evasión en comparación con pagos automáticos.
- **Género**: no se encontraron diferencias significativas entre hombres y mujeres.

### 4. Evasión por variables numéricas
- **Meses de contrato**: clientes con menor antigüedad son más propensos a cancelar.
- **Cargo mensual**: quienes pagan más caro tienden a evadir en mayor proporción.
- **Cargo total**: clientes con bajo gasto acumulado muestran más evasión, lo que refuerza la idea de que cancelan temprano en su ciclo.

---

## 🔹 Conclusiones e Insights

1. **Contratos cortos = más evasión**  
   Los clientes con contrato mensual muestran una clara propensión a cancelar.  

2. **Métodos de pago influyen**  
   El pago por *“Electronic check”* se asocia con mayor churn, posiblemente por ser menos cómodo que los pagos automáticos.  

3. **Clientes nuevos son más vulnerables**  
   La mayoría de los que cancelan tienen pocos meses de contrato.  

4. **Altos cargos mensuales = mayor churn**  
   El precio elevado es un factor que impulsa la evasión.  

---

## 🔹 Recomendaciones

1. **Incentivar contratos de largo plazo**  
   Ofrecer descuentos y beneficios para clientes que cambien de “month-to-month” a “one year” o “two year”.

2. **Fomentar pagos automáticos**  
   Dar beneficios a clientes que pasen de *Electronic check* a *Bank transfer* o *Credit card (automatic)*.  

3. **Programas de fidelización para clientes nuevos**  
   Primeros 6 meses son críticos: ofrecer promociones, soporte especial y beneficios exclusivos para reducir cancelaciones tempranas.  

4. **Revisión de precios y segmentación**  
   Analizar clientes con cargos mensuales altos y ofrecer planes más flexibles para mejorar la retención.  

---

## 🚀 Conclusión General

Este análisis permitió **limpiar, transformar y explorar los datos de clientes de Telecom X**, identificando patrones claros en la evasión.  
Con los insights y recomendaciones, el equipo de Data Science podrá avanzar en modelos predictivos que anticipen la fuga de clientes y en estrategias de negocio para **mejorar la retención y reducir la tasa de churn**.
